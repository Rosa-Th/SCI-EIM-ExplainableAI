## Data plotting notebook (Plotting Results from 4a-4c)

```python
from matplotlib import pyplot as plt
import numpy as np

"""xgboost_acc = 0.680365296803653
xgboost_f1 = 0.6790551296302999
xgboost_roc = 0.8183944198121308

mlp_acc = 0.6164383561643836
mlp_f1 = 0.6133805040898491
mlp_roc = 0.7326575530947196

sequantial_acc = 0.6666666666666666
sequantial_f1 = 0.6684647477821015
sequantial_roc = 0.8151189053000794"""

# Data for plotting
metrics = ['Accuracy', 'F1 Score', 'ROC AUC']
xgboost_scores = [0.680365296803653, 0.6790551296302999, 0.8183944198121308]
mlp_scores = [0.634703196347032, 0.6316526206554497, 0.7376978926166281]
sequential_scores = [0.6666666666666666, 0.6684647477821015, 0.8151189053000794]

# Convert scores to percentages
xgboost_scores = [x * 100 for x in xgboost_scores]
mlp_scores = [x * 100 for x in mlp_scores]
sequential_scores = [x * 100 for x in sequential_scores]

# Define bar width and positions
bar_width = 0.2
r1 = np.arange(len(metrics))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Plotting the bars
plt.figure(figsize=(15, 10))
plt.bar(r1, xgboost_scores, color='lightblue', width=bar_width, edgecolor='grey', label='XGBoost')
plt.bar(r2, mlp_scores, color='lightgreen', width=bar_width, edgecolor='grey', label='MLP')
plt.bar(r3, sequential_scores, color='pink', width=bar_width, edgecolor='grey', label='Sequential')

# Adding labels and title
plt.xlabel('Metrics', fontsize=26)
plt.ylabel('Scores (%)', fontsize=26)
plt.xticks([r + bar_width for r in range(len(metrics))], metrics, fontsize=24)
plt.yticks(np.arange(50, 101, 5), fontsize=24)  # Setting the tick marks starting from 50% to 100%
plt.ylim(50, 85)  # Set the y-axis to start from 50%
plt.title('Comparison of Model Performance Metrics', fontsize=34)

# Adding legend
plt.legend(title="Model Type", fontsize=24, title_fontsize=24)

# Display plot
plt.show()
```
